# what is a Generator

- Python generators are simple way of creating iterators

```python
# iterable
class Range:
    '''Create our own range class
       for iterator == iterable_class'''
    def __init__(self,start,end):
        self.start = start
        self.end = end
        
    def __iter__(self):
        return Range_Iterator(self)
# iterator    
class Range_Iterator:
    """Create an loop for each number
       == iterator_class"""
    
    def __init__(self,iterable_obj):
        self.iterable = iterable_obj
    
    def __iter__(self):
        return self
    
    def __next__(self):
        
        if self.iterable.start >= self.iterable.end:
            raise StopIteration
            
        current = self.iterable.start
        
        self.iterable.start += 1
        
        return current
```

# Range Function using Generator

```python
def def_range(start,end):
    
    for i in range(start,end):
        yield i
```

```python
gen = def_range(26,31)

for i in gen:
    print(i)
```

```python
for i in def_range(26,31):
    print(i)
```

# represnt infinite streams

```python
def all_even():
    n = 0
    while True:
        yield n
        n += 2
```

```python
gen = all_even()

next(gen)
```

```python
next(gen)
```

```python
for i in gen:
    print(i)
    
    if i > 100:
        break
```

# Chaining Generators

```python
def fibonacci_numbers(nums):
    x,y = 0,1
    for _ in range(nums):
        x, y = y, x+y
        yield x
```

```python
def square(nums):
    for num in nums:
        yield num**2
```

```python
print(sum(square(fibonacci_numbers(10))))
```
