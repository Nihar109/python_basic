## Star pattern

```python
def pyramid(n):
    # outer loop to handle number of rows
    # n in this case
    for i in range(0,n):
        # inner loop to handle number of columns
        # values changing acc. to outer loop
        for j in range(0,i+1):
            print('* ',end='')
        
        # ending line after each row
        print()
        
pyramid(5)
```

```python
def inverse_pyramid(n):
    for i in range(0,n):
        for j in range(0,n-i):
            print('* ',end='')
        print()
        
inverse_pyramid(5)
```

```python
# pyramid for number
def pyramid(n):
    num = 1
    for i in range(0,n):
        for j in range(0,i+1):
            print(num,end=' ')
            num += 1
        print()
        
pyramid(5)
```

```python
# pyramid for number
def pyramid(n):
    for i in range(0,n):
        num = 1
        for j in range(0,i+1):
            print(num,end=' ')
            num += 1
        print()
        
pyramid(5)
```
